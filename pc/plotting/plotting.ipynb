```python
import time
%matplotlib notebook

import os
from collections import deque

import numpy as np
import matplotlib.pyplot as plt
import time

filePath = '/home/fardin/devel/ltzvisor/pc/cmake-build-debug/state.csv'
```

```python
def pltsin(ax, x, y, colors=['b']):
    ax.cla()
    for color in colors:
        ax.plot(x, y, color)

    fig.canvas.draw()
```

```python
fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,1)
ax.set_ylim(-1,1)


lastLine = '--'
time_vector = np.linspace(0,1,100)
elevation = deque([0] * 100)

while True:
    f = open(filePath,'r')
    lines = f.readlines()
    if lines[-1] != lastLine:
        lastLine = lines[-1]
        line = lastLine.split(',')
        
        elevation_sample = float(line[0])
        elevation.popleft()
        elevation.append(elevation_sample)
        
        pltsin(ax, time_vector, elevation, ['r'])        
        print(lines[-1])
        

    time.sleep(0.1)
```
